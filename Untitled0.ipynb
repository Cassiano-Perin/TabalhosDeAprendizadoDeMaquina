```
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import LocalOutlierFactor
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import make_scorer
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif, chi2
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from xgboost import plot_tree
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```
cancer = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00451/dataR2.csv",header=0)
cancer.head()
```

```
cancer.info()
```

```
cancer.duplicated().sum()
```

```
cancer.loc[cancer["Classification"]==2,"Classification"] = 0
X_treino, X_teste, y_treino, y_teste = train_test_split(cancer.drop(labels='Classification',axis=1),cancer['Classification'],test_size=0.15,random_state=41,shuffle=True,stratify=cancer['Classification'])
```

```
lof = LocalOutlierFactor()
outlier = lof.fit_predict(X_treino)
mascara = outlier != -1

print('Antes: ',X_treino.shape)
print('Depois: ',X_treino.loc[mascara,:].shape)

X_treino_ = X_treino.loc[mascara,:]
y_treino_ = y_treino[mascara]
print(y_treino_.shape)
```

```
class Transformar(BaseEstimator,TransformerMixin):
  def __init__(self,modelo = 0):
    self.modelo = modelo

  def fit(self, X, y = None):
    return self

  def transform(self, X, y = None):
    df = X.copy()

    if self.modelo == 0:
      scaler = MinMaxScaler()
      df = scaler.fit_transform(df)
    elif self.modelo == 1:
      scaler = StandardScaler()
      df = scaler.fit_transform(df)
    else:
      df = df.values

    return df
```

```
# Testando a classe Transformar
print(X_treino.head())
modelo = Transformar(modelo=1)
x_ = modelo.fit_transform(X_treino)
print(x_)
```

```
pipeline_de_preprocessamento = Pipeline(steps=[('transformar',Transformar())])
```

```
# Testar pipeline de preprocessamento
colunas_ = X_treino.columns.to_list()
teste = pipeline_de_preprocessamento.fit_transform(X_treino)
teste = pd.DataFrame(teste,columns=colunas_)
teste.head()
```

```
pipe = Pipeline(steps = [('full_pipeline', pipeline_de_preprocessamento),
                         ("fs",SelectKBest()),
                         ("clf",DecisionTreeClassifier())])
esp_de_busca = [{"clf":[(DecisionTreeClassifier())],
                 "clf__criterion": ["gini","entropy"], 
                 "clf__splitter":["best"],
                 "fs__k":[3,6,9],
                 "fs__score_func":[f_classif, mutual_info_classif, chi2],
                 "full_pipeline__transformar__modelo":[0,1,2]},
                {"clf":[(KNeighborsClassifier())],
                 "clf__n_neighbors": [3,5,7], 
                 "clf__algorithm":["auto","ball_tree", "kd_tree", "brute"],
                 "clf__p":[1,2],
                 "fs__k":[3,6,9],
                 "fs__score_func":[f_classif, mutual_info_classif, chi2],
                 "full_pipeline__transformar__modelo":[0,1,2]},
                {"clf":[(RandomForestClassifier())],
                 "clf__n_estimators": [40,80,120],
                 "clf__criterion": ["gini","entropy"],
                 "clf__max_leaf_nodes": [4,16,32,64],
                 "fs__k":[3,6,9],
                 "fs__score_func":[f_classif, mutual_info_classif, chi2],
                 "full_pipeline__transformar__modelo":[0,1,2]},
                {"clf":[(XGBClassifier())],
                 "clf__n_estimators": [40,80,120],
                 "clf__max_depth": [2,4,6], 
                 "clf__learning_rate": [0.1],
                 "clf__subsample": [0.3,0.6,0.9],
                 "clf__colsample_bytree": [0.2], 
                 "clf__tree_method":["auto"],
                 "fs__k":[3,6,9],
                 "fs__score_func":[f_classif, mutual_info_classif, chi2],
                 "full_pipeline__transformar__modelo":[0,1,2]}]
kfold = StratifiedKFold(n_splits=10,random_state=15,shuffle=True)
scoring = {'AUC': 'roc_auc', 'Accuracy': make_scorer(accuracy_score)}
grade = RandomizedSearchCV(estimator=pipe, 
                    param_distributions=esp_de_busca,
                    n_iter=1872,
                    cv=kfold,
                    scoring=scoring,
                    return_train_score=True,
                    n_jobs=-1,
                    refit="Accuracy")
todos_modelos = grade.fit(X_treino,y_treino)
```

```
print('Melhor resultado = ',todos_modelos.best_score_)
print('Melhores parametros = ',todos_modelos.best_params_)
```

```
resultados = pd.DataFrame(todos_modelos.cv_results_)
resultados
```

```
resultados_ = resultados[['mean_train_AUC', 'std_train_AUC','mean_test_AUC', 'std_test_AUC',"rank_test_AUC"]].copy()
for nome in resultados_.columns:
  resultados_.rename(columns={nome:nome.replace("test","validation")}, inplace=True)
resultados_
```

```
pred = todos_modelos.predict(X_teste)
```

```
confusion_matrix(pred,y_teste,
                 labels=[1,0])
```

```
print(accuracy_score(y_teste, pred))
print(classification_report(y_teste,pred))
```

```
fig, ax = plt.subplots(1,1,figsize=(7,4))

ConfusionMatrixDisplay(confusion_matrix(pred,y_teste,labels=[1,0])).plot(values_format=".0f",ax=ax)

ax.set_xlabel("True Label")
ax.set_ylabel("Predicted Label")
plt.show()
```
