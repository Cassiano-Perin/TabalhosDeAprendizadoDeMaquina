```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy
```

```
cancer = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00451/dataR2.csv",header=0)
cancer.loc[cancer["Classification"]==2,"Classification"] = 0
X_treino_1, X_teste_1, y_treino_1, y_teste_1 = train_test_split(cancer.drop(labels='Classification',axis=1),cancer['Classification'],test_size=0.15,random_state=41,shuffle=True,stratify=cancer['Classification'])
norma = StandardScaler()
X_treino_1 = norma.fit_transform(X_treino_1)
X_teste_1 = norma.transform(X_teste_1)
```

```
tf.random.set_seed(3)

modelo = tf.keras.Sequential([
                             tf.keras.layers.Dense(9, activation=tf.nn.tanh),
                             tf.keras.layers.Dense(15, activation=tf.nn.tanh),
                             tf.keras.layers.Dense(1, activation = tf.nn.sigmoid)
                             ])

loss = tf.keras.losses.BinaryCrossentropy()

accuracy = tf.keras.metrics.BinaryAccuracy()

optimizer = tf.keras.optimizers.Adam(learning_rate=0.009265680840246284)

modelo.compile(optimizer=optimizer, loss=loss, metrics=[accuracy])

history = modelo.fit(x=numpy.array(X_treino_1),
                    y=numpy.array(y_treino_1),
                    epochs=50,
                    validation_data=(numpy.array(X_teste_1),numpy.array(y_teste_1)),
                    verbose=0)
```

```
plt.plot(history.history['binary_accuracy'])
plt.plot(history.history['val_binary_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
```

```
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
```

```
loss, acc = modelo.evaluate(x=X_treino_1,y=y_treino_1)
print('Train loss: %.3f - acc: %.3f' % (loss, acc))

loss_, acc_ = modelo.evaluate(x=X_teste_1,y=y_teste_1, batch_size=32)
print('Test loss: %.3f - acc: %.3f' % (loss_, acc_))
```
