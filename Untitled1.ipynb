```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
import kerastuner as kt
import matplotlib.pyplot as plt
import numpy as np
import h5py
import time
import datetime
import pytz
import IPython
from kerastuner.tuners import RandomSearch
```

```
cancer = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00451/dataR2.csv",header=0)
cancer.head()
```

```
cancer.info()
```

```
cancer.duplicated().sum()
```

```
cancer.loc[cancer["Classification"]==2,"Classification"] = 0
cancer_ = cancer.loc[:,['Glucose','Resistin','Age','BMI','HOMA','Classification']]
X_treino_0, X_teste_0, y_treino_0, y_teste_0 = train_test_split(cancer_.drop(labels='Classification',axis=1),cancer_['Classification'],test_size=0.15,random_state=41,shuffle=True,stratify=cancer_['Classification'])
X_treino_1, X_teste_1, y_treino_1, y_teste_1 = train_test_split(cancer.drop(labels='Classification',axis=1),cancer['Classification'],test_size=0.15,random_state=41,shuffle=True,stratify=cancer['Classification'])
```

```
lof = LocalOutlierFactor()
outlier = lof.fit_predict(X_treino_0)
mascara = outlier != -1

print('Antes: ',X_treino_0.shape)
print('Depois: ',X_treino_0.loc[mascara,:].shape)

X_treino_2 = X_treino_0.loc[mascara,:]
y_treino_2 = y_treino_0[mascara]
print(y_treino_2.shape)
```

```
lof = LocalOutlierFactor()
outlier = lof.fit_predict(X_treino_1)
mascara = outlier != -1

print('Antes: ',X_treino_1.shape)
print('Depois: ',X_treino_1.loc[mascara,:].shape)

X_treino_3 = X_treino_1.loc[mascara,:]
y_treino_3 = y_treino_1.loc[mascara]
print(y_treino_3.shape)
```

```
aux_0 = X_teste_0
aux_1 = X_teste_1
norma = StandardScaler()
X_treino_0 = norma.fit_transform(X_treino_0)
X_teste_0 = norma.transform(X_teste_0)
X_treino_1 = norma.fit_transform(X_treino_1)
X_teste_1 = norma.transform(X_teste_1)
X_treino_2 = norma.fit_transform(X_treino_2)
X_teste_2 = norma.transform(aux_0)
X_treino_3 = norma.fit_transform(X_treino_3)
X_teste_3 = norma.transform(aux_1)
```

```
def modelos(hp):
  modelo = tf.keras.Sequential()
  modelo.add(tf.keras.layers.Dense(units=hp.Int('units',min_value = 3,max_value = 21,step = 1),
               activation= hp.Choice('dense_activation',values=['relu', 'tanh'],default='relu'),kernel_initializer="he_uniform",dtype='float64'))
  modelo.add(tf.keras.layers.Dense(1,activation=tf.nn.sigmoid,kernel_initializer="he_uniform",dtype='float64'))
  perda = tf.keras.losses.BinaryCrossentropy()
  acuracia = tf.keras.metrics.BinaryAccuracy()
  otimiza = hp.Choice('optimizer',['adam','SGD'])
  if otimiza == 'adam':
    opt = tf.keras.optimizers.Adam(learning_rate=hp.Float('lrate_adam',min_value=1e-4,max_value=1e-2,sampling='LOG'))
  else:
    opt = tf.keras.optimizers.SGD(learning_rate=hp.Float('lrate_sgd',min_value=1e-4,max_value=1e-2,sampling='LOG'))
  modelo.compile(optimizer=opt, loss=perda, metrics=[acuracia])
  return modelo
```

```
LOG_DIR = f"{int(time.time())}"
```

```
class ClearTrainingOutput(tf.keras.callbacks.Callback):
  def on_train_end(*args, **kwargs):
    IPython.display.clear_output(wait = True)
```

```
turne0 = RandomSearch(modelos,
                     objective='val_binary_accuracy',
                     max_trials=114,  
                     executions_per_trial=4, 
                     directory=LOG_DIR)
turne1 = RandomSearch(modelos,
                     objective='val_binary_accuracy',
                     max_trials=114,  
                     executions_per_trial=4, 
                     directory=LOG_DIR)
turne2 = RandomSearch(modelos,
                     objective='val_binary_accuracy',
                     max_trials=114,  
                     executions_per_trial=4, 
                     directory=LOG_DIR)

turne3 = RandomSearch(modelos,
                     objective='val_binary_accuracy',
                     max_trials=114,  
                     executions_per_trial=4, 
                     directory=LOG_DIR)
```

```
turne0.search(np.array(X_treino_0),
             np.array(y_treino_0), 
             epochs = 25,
             verbose=0,
             validation_data = (np.array(X_teste_0), np.array(y_teste_0)),
             callbacks = [ClearTrainingOutput()])
turne1.search(np.array(X_treino_1),
             np.array(y_treino_1), 
             epochs = 25,
             verbose=0,
             validation_data = (np.array(X_teste_1), np.array(y_teste_1)),
             callbacks = [ClearTrainingOutput()])
turne2.search(np.array(X_treino_2),
             np.array(y_treino_2), 
             epochs = 25,
             verbose=0,
             validation_data = (np.array(X_teste_2), np.array(y_teste_0)),
             callbacks = [ClearTrainingOutput()])
turne3.search(np.array(X_treino_3),
             np.array(y_treino_3), 
             epochs = 25,
             verbose=0,
             validation_data = (np.array(X_teste_3), np.array(y_teste_1)),
             callbacks = [ClearTrainingOutput()])
```

```
print('caso 0:')
print(turne0.results_summary())
print('caso 1:')
print(turne1.results_summary())
print('caso 2:')
print(turne2.results_summary())
print('caso 3:')
print(turne3.results_summary())
```

```
print('Caso 0:')
print(turne0.get_best_hyperparameters()[0].values)
print('Caso 1:')
print(turne1.get_best_hyperparameters()[0].values)
print('Caso 2:')
print(turne2.get_best_hyperparameters()[0].values)
print('Caso 3:')
print(turne3.get_best_hyperparameters()[0].values)
```

```
best0 = turne0.get_best_hyperparameters()[0]
modelo0 = turne0.hypermodel.build(best0)
modelo0.fit(np.array(X_treino_0), np.array(y_treino_0), epochs = 28, validation_data = (np.array(X_teste_0), np.array(y_teste_0)),verbose=0)

best1 = turne1.get_best_hyperparameters()[0]
modelo1 = turne1.hypermodel.build(best1)
modelo1.fit(np.array(X_treino_1), np.array(y_treino_1), epochs = 28, validation_data = (np.array(X_teste_1), np.array(y_teste_1)),verbose=0)

best2 = turne2.get_best_hyperparameters()[0]
modelo2 = turne2.hypermodel.build(best2)
modelo2.fit(np.array(X_treino_2), np.array(y_treino_2), epochs = 28, validation_data = (np.array(X_teste_2), np.array(y_teste_0)),verbose=0)

best3 = turne3.get_best_hyperparameters()[0]
modelo3 = turne3.hypermodel.build(best3)
modelo3.fit(np.array(X_treino_3), np.array(y_treino_3), epochs = 28, validation_data = (np.array(X_teste_3), np.array(y_teste_1)),verbose=0)
```

```
print('caso 0:')
loss_0, acc_0 = modelo0.evaluate(x=np.array(X_teste_0),y=np.array(y_teste_0))
print('Test loss: %.3f - acc: %.3f' % (loss_0, acc_0))

print('caso 1:')
loss_1, acc_1 = modelo1.evaluate(x=np.array(X_teste_1),y=np.array(y_teste_1))
print('Test loss: %.3f - acc: %.3f' % (loss_1, acc_1))

print('caso 2:')
loss_2, acc_2 = modelo2.evaluate(x=np.array(X_teste_2),y=np.array(y_teste_0))
print('Test loss: %.3f - acc: %.3f' % (loss_2, acc_2))

print('caso 3:')
loss_3, acc_3 = modelo3.evaluate(x=np.array(X_teste_3),y=np.array(y_teste_1))
print('Test loss: %.3f - acc: %.3f' % (loss_3, acc_3))
```

```
def modelos_complementares(hp):
  modelo = tf.keras.Sequential()
  modelo.add(tf.keras.layers.Dense(units=hp.Int('units_1',min_value = 3,max_value = 21,step = 1),
               activation= hp.Choice('dense_activation_1',values=['relu', 'tanh'],default='relu'),kernel_initializer="he_uniform",dtype='float64'))
  modelo.add(tf.keras.layers.Dense(units=hp.Int('units_2',min_value = 3,max_value = 21,step = 1),
               activation= hp.Choice('dense_activation_2',values=['relu', 'tanh'],default='relu'),kernel_initializer="he_uniform",dtype='float64'))
  modelo.add(tf.keras.layers.Dense(1,activation=tf.nn.sigmoid,kernel_initializer="he_uniform",dtype='float64'))
  perda = tf.keras.losses.BinaryCrossentropy()
  acuracia = tf.keras.metrics.BinaryAccuracy()
  otimiza = hp.Choice('optimizer',['adam','SGD'])
  if otimiza == 'adam':
    opt = tf.keras.optimizers.Adam(learning_rate=hp.Float('lrate_adam',min_value=1e-4,max_value=1e-2,sampling='LOG'))
  else:
    opt = tf.keras.optimizers.SGD(learning_rate=hp.Float('lrate_sgd',min_value=1e-4,max_value=1e-2,sampling='LOG'))
  modelo.compile(optimizer=opt, loss=perda, metrics=[acuracia])
  return modelo
```

```
LOG_DIR = f"{int(time.time())}"
```

```
turne_0 = RandomSearch(modelos_complementares,
                     objective='val_binary_accuracy',
                     max_trials=114,  
                     executions_per_trial=4, 
                     directory=LOG_DIR)
turne_1 = RandomSearch(modelos_complementares,
                     objective='val_binary_accuracy',
                     max_trials=114,  
                     executions_per_trial=4, 
                     directory=LOG_DIR)
turne_2 = RandomSearch(modelos_complementares,
                     objective='val_binary_accuracy',
                     max_trials=114,  
                     executions_per_trial=4, 
                     directory=LOG_DIR)

turne_3 = RandomSearch(modelos_complementares,
                     objective='val_binary_accuracy',
                     max_trials=114,  
                     executions_per_trial=4, 
                     directory=LOG_DIR)
```

```
turne_0.search(np.array(X_treino_0),
             np.array(y_treino_0), 
             epochs = 25,
             verbose=0,
             validation_data = (np.array(X_teste_0), np.array(y_teste_0)),
             callbacks = [ClearTrainingOutput()])
turne_1.search(np.array(X_treino_1),
             np.array(y_treino_1), 
             epochs = 25,
             verbose=0,
             validation_data = (np.array(X_teste_1), np.array(y_teste_1)),
             callbacks = [ClearTrainingOutput()])
turne_2.search(np.array(X_treino_2),
             np.array(y_treino_2), 
             epochs = 25,
             verbose=0,
             validation_data = (np.array(X_teste_2), np.array(y_teste_0)),
             callbacks = [ClearTrainingOutput()])
turne_3.search(np.array(X_treino_3),
             np.array(y_treino_3), 
             epochs = 25,
             verbose=0,
             validation_data = (np.array(X_teste_3), np.array(y_teste_1)),
             callbacks = [ClearTrainingOutput()])
```
